```python
import pandas as pd
import numpy as np
import pprint
import operator
import re

pd.options.display.max_columns = None
```

```python
df = pd.read_csv('./jdp_2017.csv',low_memory=False)
df.shape
```

```python
#df.head()
```

```python
temp = df['Make_Series']
df = df.filter(like='Verbatim',axis=1)
df.shape
```

```python
df = df.dropna(axis=1,how='all')
df = df.dropna(axis=0,how='all')
df.shape
```

```python
df.insert(loc=0, column='Make_Series', value=temp)
df.head()
```

```python
count = 0
makes = []
series = []
verbatims = []

df = df.reset_index(drop=True)
#df.to_csv('temp.csv')


for i in range(len(df)):
    for j in range(1,len(df.iloc[i])):
        if pd.notnull(df.iloc[i,j]) and df.iloc[i,j]!=0.0 and df.iloc!=1.0 and df.iloc[i,j]!=0 and df.iloc[i,j]!=1:
            count += 1
            
            if df.iloc[i,0][0]=='B':
                makes.append('BMW')
                series.append(df.iloc[i,0][4:])
            elif df.iloc[i,0][0]=='M':
                makes.append('MINI')
                series.append(df.iloc[i,0][5:])

            verbatims.append(df.iloc[i,j])
            
print(count)           
```

```python
new_df = pd.DataFrame(columns=['Make','Series','Verbatim'])
new_df['Make'] = pd.Series(makes)
new_df['Series'] = pd.Series(series)
new_df['Verbatim'] = pd.Series(verbatims)
print(new_df.shape)
new_df.head()
```

```python
#new_new_df = pd.concat([bmw_df,mini_df],ignore_index=True)
new_df = new_df.fillna("")
new_df['Verbatim'] = new_df['Verbatim'].replace(to_replace='\(([^)]+)\)',value='',regex=True)
print(new_df.shape)
new_df.head()
```

```python
import nltk
nltk.download('subjectivity')
nltk.download('vader_lexicon')
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import subjectivity
from nltk.sentiment import SentimentAnalyzer
from nltk.sentiment.util import *
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
verbatim = new_df['Verbatim'].tolist()
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
i = 0
makes = []
series = []
statements = []
scores = []

for entry in verbatim:
    #print("{0}".format(new_df.loc[i,'Model and Type']))
    #print("Total Problems: {0}, Category: {1}, Mode: {2}".format(new_df.loc[i,'Total Problems'],new_df.loc[i,'Category'],new_df.loc[i,'Mode']))
    #print(entry)
    ss = sid.polarity_scores(entry)
    j = 0
    for k in sorted(ss):
        #print('{0}: {1}, '.format(k,ss[k]),end='')
        if j == 0:
            scores.append(ss[k])
            j+=1
    #print("\n")
    
    makes.append(new_df.loc[i,'Make'])
    series.append(new_df.loc[i,'Series'])
    statements.append(entry)
    
    i+=1
```

```python
stuff = [('Make',makes), ('Series',series), ('Verbatim',statements), ('Score',scores)]
new_new_df = pd.DataFrame.from_items(stuff)
new_new_df.head()
```

```python
new_new_df.to_csv('2017_sentiments.csv')
```
