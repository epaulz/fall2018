```python
import pandas as pd
import numpy as np
import pprint
import operator
import re

pd.options.display.max_columns = None
```

```python
df15 = pd.read_csv('./2015_sentiments.csv')
df16 = pd.read_csv('./2016_sentiments.csv')
df17 = pd.read_csv('./2017_sentiments.csv')
df18 = pd.read_csv('./2018_sentiments.csv')
```

```python
print(df15.shape)
print(df16.shape)
print(df17.shape)
print(df18.shape)
```

```python
df_list = [df15,df16,df17,df18]
all_make = []
all_series = []

for df in df_list:
    for i in range(len(df)):
        if df.loc[i,'Series'] not in all_series:
            all_make.append(df.loc[i,'Make'])
            all_series.append(df.loc[i,'Series'])
            
#for x in all_vehicles:
#    print(x)
```

```python
new_df = pd.DataFrame(columns=['Year','Make','Series','Score_Count','Score_Avg'])
i=0

for j in range(len(all_series)):
    yr = 2015
    
    for df in df_list:
        if len(df[df['Series']==all_series[j]]) > 0:
            new_df.loc[i] = [yr,all_make[j],all_series[j],len(df[df['Series']==all_series[j]]),df[df['Series']==all_series[j]]['Score'].mean()]
            
        i+=1
        yr+=1

new_df = new_df.sort_values(by=['Year'])
```

```python
new_df['Year'] = new_df['Year'].astype('int')
new_df['Score_Count'] = new_df['Score_Count'].astype('int')
```
